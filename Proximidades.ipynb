## **Medidas de Proximidad**

### Miden el grado de semejanza (o de desemejanza) que presentan dos elementos $(i,j)$, a causa de sus caracteristicas $d(i,j)$. Las m茅tricas de disimilaridad, aumentan de valor al decrecer la semejanza. Del mismo modo, hay m茅tricas de similaridad que aumentan de valor al crecer la semanaja y viceversa.

## **Medidas de Disimilaridad**

### De las medidas de distancia se f贸rmulan las siguientes propiedades:

* ### $(a)\quad d(i,j) \geq 0, \,\forall i,j$
* ### $(b)\quad d(i,i) = 0, \,\forall i$ 
* ### $(c)\quad d(i,j) = d(j,i), \,\forall i,j$
* ### $(d)\quad d(i,j) \leq d(i,k) + d(k,j), \,\forall i,j,k$

```python
# Dependencias de la funci贸n
from typing import Literal
from scipy.spatial.distance import pdist, squareform
import pandas as pd
import numpy as np
import pyreadstat # Est谩 no es una dependencia como tal pero
# me sirve puesto que los datos provienen de un archivo SPSS

# Definimos una funcion
def distances(
    data: pd.DataFrame,
    kind: Literal[
        "braycurtis", "canberra", "chebyshev", "cityblock", "correlation", "cosine", "dice",
        "euclidean", "hamming", "jaccard", "jensenshannon", "kulczynski1", "mahalanobis",
        "matching", "minkowski", "rogerstanimoto", "russellrao", "seuclidean", "sokalmichener",
        "sokalsneath", "sqeuclidean", "yule"
    ],
    p: int = 0
):

    # Una lista con los tipos de distancia posibles, mismos que de la funci贸n "pdist" de scipy
    allowed_kinds = [
        "braycurtis", "canberra", "chebyshev", "cityblock", "correlation", "cosine", "dice",
        "euclidean", "hamming", "jaccard", "jensenshannon", "kulczynski1", "mahalanobis",
        "matching", "minkowski", "rogerstanimoto", "russellrao", "seuclidean", "sokalmichener",
        "sokalsneath", "sqeuclidean", "yule"
    ]

    # Comprobar si kind est谩 dentro de los permitidos
    if kind not in allowed_kinds:
        raise ValueError(f"Tipo inv谩lido, de la lista: {allowed_kinds}, se escribi贸 '{kind}' en su lugar.")
    if not isinstance(kind, str):
        raise TypeError("kind debe ser un string.")

    # Comprobar si p es igual a cero (que es por defecto)
    if p == 0: 
        distance_with_zero = pdist(X= data, 
                     metric = kind,
                    )
        matrix_with_zero = squareform(distance_with_zero)
        df_with_zero = pd.DataFrame(matrix_with_zero,
                                index=data.index,
                                columns= data.index)
        # Para que el DataFrame no tenga repetici贸n de nombres
        df_with_zero.index.name = None

        return df_with_zero
    
    # Caso contrario
    else: 
        distance = pdist(X= data, 
                     metric = kind,
                    p = p)
        matrix = squareform(distance)
        df = pd.DataFrame(matrix,
                                index=data.index,
                                columns= data.index)
        # Para que el DataFrame no tenga repetici贸n de nombres
        df.index.name = None
        
        return df
    
```

## **Distancia Eucl铆deana**

### La distancia Euclidiana se utiliza para medir la cercan铆a o disimilaridad entre dos puntos en un espacio -dimensional. Es una m茅trica fundamental en matem谩ticas, estad铆stica y aprendizaje autom谩tico porque proporciona una manera simple e intuitiva de cuantificar qu茅 tan diferentes son dos objetos en funci贸n de sus caracter铆sticas.
$$
\LARGE
d_{ij} = \sqrt{\sum_{k=1}^n (X_{ik} - X_{jk})^2}

```python
euclidean = distances(data = df,
                        kind = "euclidean")
euclidean.round(3)
```

## **Distancia Minkowski**

### Es una generalizaci贸n de varias m茅tricas de distancia, incluyendo la distancia Euclidiana y la Manhattan. Permite ajustar la m茅trica seg煤n el par谩metro , lo que la hace muy flexible para medir disimilaridad en diferentes contextos.

$$
\LARGE
d(i,j) = (\sum_{h=1}^p | x_{ih} - x _{jh}|^q)^{1/q}
$$

### D贸nde:

* ### $q$: se obtiene la **Distancia Manhattan** o de **City-Block**.
* ### $q$: se obtiene la **Distancia Eucl铆dea**
* ### $q \to \infty$: se obtiene la **Distancia de Chebychev**.

```python
p_values = [1, 2, 40]

distances_within_minkowski = ["Manhattan", "euclidean", "Chebychev"]

for distance, p in zip(distances_within_minkowski, p_values):
    minkowski = distances(data=df, kind="minkowski", p=p)
    print(f"Distancia {distance} por Minkowski con un p = {p}:\n")
    print(minkowski.round(3))
    print("\n" + "-"*80 + "\n")
```

## **Distancia Mahalanobis**

### La distancia de Mahalanobis es una m茅trica que mide la distancia entre un punto y una distribuci贸n en un espacio multidimensional. Es particularmente 煤til cuando las variables est谩n correlacionadas y tienen diferentes escalas, ya que ajusta las diferencias teniendo en cuenta la matriz de covarianza.


$$
\LARGE
d(i,j) = D^2 = (x_i - x_j)' \sum^{-1} (x_i - x_j)
$$

```python
mahalanobis = distances(data = df,
                        kind = "mahalanobis")
mahalanobis.round(3)
```

## **Coeficiente de Similiaridad Gower** 

### El coeficiente de Gower es una medida de similitud especialmente dise帽ada para trabajar con conjuntos de datos que contienen diferentes tipos de variables, como num茅ricas, binarias (s铆/no) y categ贸ricas (por ejemplo, color, g茅nero). Esto lo convierte en una herramienta invaluable en el an谩lisis de datos cuando se enfrentan a conjuntos de datos heterog茅neos, que son muy comunes en la pr谩ctica.

$$
\LARGE
d_{ij} = \frac{\sum_{k=1}^P W_k\,  \delta_{ijk} \, S_{ijk}}{\sum_{k=1}^P W_k \, \delta_{ikj}}
$$

```python
# Creamos unos datos
# Los datos son de: https://pypi.org/project/gower/
df3=pd.DataFrame({'age':[21,21,19, 30,21,21,19,30],
'gender':['M','M','N','M','F','F','F','F'],
'salary':[3000.0,1200.0 ,32000.0,1800.0 ,2900.0 ,1100.0 ,10000.0,1500.0],
'available_credit':[2200,100,22000,1100,2000,100,6000,2200]})
df3["id"] = range(1, len(df3) + 1)
df3.set_index("id", inplace = True)

df3
```

```python
# Dependencia de la funci贸n
import numpy
import gower
import pandas as pd

# Definimos una funci贸n
def gower_similarity(data: pd.DataFrame):

    # Hallamos los coeficientes de gower
    gower_coeficients = gower.gower_matrix(data)

    # Lo volvemos un dataframe porque el gower_matrix regresa un array
    gower_coeficients_df = pd.DataFrame(gower_coeficients,
                                        index = data.index,
                                        columns = data.index
                                        )
    
    # Ponemos "None" para que no repita el indetificativo,
    # ya que se intuye por los datos
    gower_coeficients_df.index.name = None

    # Devolvemos 
    return gower_coeficients_df

# probamos la funci贸n

gower_coeficients = 1 - gower_similarity(df3) # 1 - para que sea una similitud

gower_coeficients.round(3)
```

## **Medidas de relaci贸n lineal ( $Y = a + bX$)**

### **Coeficiente de correlaci贸n lineal:**

## **Medidas basadas en relaci贸n: $Y=bX$**

### **Coseno entre variables:**

### Es una medida de similitud entre dos vectores no nulos en un espacio vectorial. Conceptualmente, mide el coseno del 谩ngulo entre estos dos vectores. Cuanto m谩s peque帽o sea el 谩ngulo, m谩s similares ser谩n los vectores.

$$
\LARGE
d(X_i, X_j) = cos(\overline{X_i}, \overline{X_j})
$$

```python
from scipy.spatial.distance import cosine

simil_cosine = 1 - cosine(df2["atletism"], df2["calculo"])

simil_cosine
```

## **Coeficiente de Correlaci贸n de Spearman**

### Es una medida estad铆stica que nos indica la fuerza y la direcci贸n de la relaci贸n entre dos variables. A diferencia del coeficiente de Pearson, que asume una distribuci贸n normal de los datos, Spearman no tiene esta restricci贸n. Esto lo hace especialmente 煤til cuando los datos no siguen una distribuci贸n normal o cuando estamos trabajando con variables ordinales (es decir, variables que expresan un orden, como el puesto en una carrera).

$$
\LARGE
\rho_s = 1 - \frac{6 \sum D_{i}^2}{N(N^2-1)}
$$

```python
spearman = df2.corr(method= "spearman").round(3)
spearman
```
